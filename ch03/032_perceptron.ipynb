```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 하이퍼볼릭 탄젠트 함수
def tanh(x):
    return list(map(lambda x : math.tanh(x), x))

# relu 함수
def relu(x):
    result = []
    for ele in x:
        if(ele <= 0):
            result.append(0)
        else:
            result.append(ele)

    return result

```

```python
# 시그모이드 함수 그려보기
x = np.linspace(-4, 4, 100)
sig = sigmoid(x)    # 함수를 변경하여 다른 함수도 그려보세요
plt.plot(x, sig)
plt.show()
```

```python
# 다른 함수 그려보기
x = np.linspace(-4, 4, 100)
tan_h = tanh(x)    # 함수를 변경하여 다른 함수도 그려보세요
relu_d = relu(x)    # 함수를 변경하여 다른 함수도 그려보세요

plt.figure(figsize = (15, 5))
plt.subplot(1, 2, 1)
plt.plot(x, tan_h); plt.title('tanh', fontsize = 30)
plt.subplot(1, 2, 2)
plt.plot(x, relu_d); plt.title('relu', fontsize = 30)
plt.show()
```

```python
# p.81 경사하강법 실험하기

import numpy as np
import matplotlib.pyplot as plt

lr_list = [0.001, 0.1, 0.5, 0.9]    # 여러 가지 학습률을 사용하여 값의 변화를 관찰해봅니다.

def get_derivative(lr):

    w_old = 2
    derivative = [w_old]

    y = [w_old ** 2]    # 손실 함수를 y = x^2로 정의합니다.

    for i in range(1, 10):
        # 먼저 해당 위치에서 미분값을 구합니다.
        dev_value = w_old * 2

        # 의의 값을 이용하여 가중치를 업데이트합니다.
        w_new = w_old - lr * dev_value
        w_old = w_new

        derivative.append(w_old)    # 업데이트 된 가중치를 저장합니다.
        y.append(w_old ** 2)        # 업데이트 된 가중치의 손실 값을 저장합니다.

    return derivative, y

x = np.linspace(-2, 2, 50)  # -2~2의 범위를 50구간으로 나눈 배열을 반환합니다.
x_square = [i ** 2 for i in x]

fig = plt.figure(figsize = (12, 7))

for i, lr in enumerate(lr_list):
    derivative, y = get_derivative(lr)
    ax = fig.add_subplot(2, 2, i + 1)
    ax.scatter(derivative, y, color = 'red')
    ax.plot(x, x_square)
    ax.title.set_text('lr = ' + str(lr))

plt.show()
```
